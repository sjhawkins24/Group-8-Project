```python
#Defining the function 
def data_cleaning(data):
    data["Team"] = data["Team"].str.replace(r"\d+", "", regex=True)
    data["Team"] = data["Team"].str.replace(")", "")
    data["Team"] = data["Team"].str.replace("(", "")
    data[["Abreviation", "Team"]] = data["Team"].str.split(" ", n = 1,  expand=True)
    data = data[["RK", "Team"]]
    return(data)
```

```python
n = 16
weeks = [i for i in range(1, n + 1)]
print(weeks)
years = ["2024", 
          "2023", 
          "2022", 
          "2021", 
          "2020"]
```

```python
all_data = pd.DataFrame()
for year in years: 
    for week in weeks: 
        if (year == "2020" and week == 2): 
            print("No data available")
        else:
            print(f"grabbing week {week} from {year}")
            if(week != 16):
                data = pd.read_html(f"https://www.espn.com/college-football/rankings/_/week/{week}/year/{year}/seasontype/2", 
                       match="RK")[1]
            else: 
                 data = pd.read_html(f"https://www.espn.com/college-football/rankings/_/week/1/year/{year}/seasontype/3", 
                       match="RK")[1]   
            data = data_cleaning(data)
            data["Week"] = week
            data["Year"] = year
            all_data = pd.concat([all_data, data], axis = 0, ignore_index = True)


all_data.head()
```

```python
#Write to csv 
all_data.to_csv("WeekByWeekRankings2020to2024.csv", index = False)
```
